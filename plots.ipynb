```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
attention = json.load(open('train_progress/attention.json'))
attention = np.array(attention)

attention_bl = []
attention_bl += json.load(open('train_progress/attention_baseline_1.json'))
attention_bl += json.load(open('train_progress/attention_baseline_2.json'))
attention_bl = np.array(attention_bl)

performer = []
performer += json.load(open('train_progress/performer1.json'))
performer += json.load(open('train_progress/performer2.json'))
performer += json.load(open('train_progress/performer3.json'))
performer = np.array(performer)
```

```python
fig, ax = plt.subplots()
ax.plot(attention[:,1], attention[:,2], label="Classical attention")
ax.plot(performer[:,1], performer[:,2], label="Performer")
#plt.yscale("log")
ax.set_ylim((100, 250))


def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(value/1000)
    if N == 0:
        return r"${0}$".format(N)
    else:
        return r"${0}k$".format(N)
def format_func2(value, tick_number):
    # find number of multiples of pi/2
    N = int(value)
    if N == 0:
        return r"${0}$".format(N)
    else:
        return r"${0}$".format(N)

ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
ax.yaxis.set_major_formatter(plt.FuncFormatter(format_func2))

ax.set_xlabel("Training step")
ax.set_ylabel("Perplexity")
ax.set_title("Training perplexity comparison")
ax.grid()
ax.legend()
plt.savefig('train.pdf')
```

```python
fig, ax = plt.subplots()
ax.plot(attention_bl[:,1], attention_bl[:,2], label="Baseline attention")
#plt.yscale("log")
ax.set_ylim((20, 250))

ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
ax.yaxis.set_major_formatter(plt.FuncFormatter(format_func2))

ax.set_xlabel("Training step")
ax.set_ylabel("Perplexity")
ax.set_title("Training perplexity comparison")
ax.grid()
ax.legend()
plt.savefig('train_baseline.pdf')
```
